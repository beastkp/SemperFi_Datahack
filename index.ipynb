```python
from bs4 import  BeautifulSoup
import requests
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

for item in urls:
    driver = webdriver.Chrome()
    #name of the url
    driver.get(urls[item])

    # Get the height of the initial page
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom of page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for new content to load
        time.sleep(1)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

        element = driver.find_element(by=By.CLASS_NAME, value="_2ujGx")
        element = element.get_attribute("innerHTML")
        main_grid = BeautifulSoup(element, 'lxml')
        
        for a in main_grid.find_all('a'):
            prices = []
            name = a.find('h2',class_ = '_2lmIw')
            if(name==None):
                continue
            
            price = a.find('div',class_ = '_18ToE')
            if(price.span == None):
                continue
            print(' '.join(name.text.split(' ')[1:]),price.span.text)
            prices.append(price.span.text[1:])

            prices = np.array(prices)
            range = (min(prices),max(prices))
            values[item] = range
    driver.quit()
```
